```
path = 'dbfs:/ispit/'


csv_name = path + 'df.csv'
nba = spark.read.csv(csv_name, header=True, inferSchema=True)
nba.show(900)
```

Ucitavam podatke sa igracima

```
csv_name = 'dbfs:/ispit_bd/' + 'df1.csv'
p_nba = spark.read.csv(csv_name, header=True, inferSchema=True)
p_nba.show(50)
```

Kreiramo TempView da bismo mogli da radimo sql upite nad njim

```
nba.createOrReplaceTempView("nba")
p_nba.createOrReplaceTempView("playerNba")
```

Ukupan broj postignutih poena za svaku sezonu

```
%sql
select season, sum(total_pts) from nba group by season order by season asc

```

Duzina sezona u mesecima

```
%sql
select season, months_between(max(date_game), min(date_game)) from nba group by season order by season asc
```

Broj playoffs utakmica po sezoni

```
%sql
select season, count(is_playoffs) as broj_utakmica_u_playoffs from nba where nba.is_playoffs == "Playoffs" group by season  order by season asc
```

Broj timova po sezoni

```
%sql
select season, count(distinct(home_team)) as team from nba group by season  order by season asc
```

Broj pobeda Njujork Knicks-a po sezonama na domacem terenu

```
%sql
select season, count(game_result) as broj_pobeda from nba where nba.game_result = "W" and nba.home_team = "NYK" group by season  order by season asc
```

Broj primljenih poena na domaces terenu za Njujork Knicks-e po sezonama

```
%sql
select season, sum(opponent_pts) as broj_primljenih_poena from nba where nba.home_team = "NYK" group by season  order by season asc
```

Maksimalan broj datih poena Njujork Knicks-a po sezonama na domacem terenu

```
%sql
select season, max(home_pts) from nba where nba.home_team == "NYK" group by season  order by season asc
```

Sezone u kojima su Njujork Knicks-i usli u playoffs

```
%sql
select distinct(season) from nba where (nba.home_team == "NYK") and (nba.is_playoffs == "Playoffs") order by season asc
```

Timovi sa najvise pobeda u gostima u sezoni 2015

```
%sql
select opponent_team, max(opponent_pts) as max_pts from nba where nba.season = "2015" group by opponent_team order by max_pts desc
```

Godine nastanka  i prestanka postojanja timova (mnogi timovi koji su postojali kada je nastala NBA liga vise ne postoje, a mnogi timovi su stvoreni kasnije)

```
%sql
select home_team, year(min(date_game)) as godina_nastanka, year(max(date_game)) as godina_zavrsetka from nba group by home_team order by godina_zavrsetka asc
```

Sezona u kojoj su imali najbolji koeficijent napada

```
%sql
select first_name, last_name, season, offense from playerNba where playerNba.offense == (select max(offense) from playerNba where last_name == "Jordan") or playerNba.offense == (select max(offense) from playerNba where last_name == "James")
```

HIPOTEZA 1
Sezone u kojima je LeBronov koeficijent efikasnosti napada manji od 5, njegov tim nije igrao playoffs

Trazimo sezone u kojime je koeficijent bio manji od 5

```
pivotPl = p_nba.select("first_name", "last_name", "season", "offense").where(p_nba.offense < 5)\
        .where(p_nba.last_name == "Jordan")
pivotPl.show()
```

Proveravamo da li su timovi igrali playoff (do 1998 je igrao u Cikagu, a posle u Wasingtonu, pa proveravamo prvo za Cikago)

```
pivot_teams = nba.select("home_team", "is_playoffs", "season").where((nba.home_team == "CHI") & ((nba.season == 1986) | (nba.season == 1995) | (nba.season == 1998)) & (nba.is_playoffs == "Playoffs"))
        
pivot_teams.show()
```

```
Hipoteza je netacna, posto vidimo da je Cikago igrao playoffs. Nema potrebe da proveravamo za Vasington, posto je hipoteza vec oborena.
```

HIPOTEZA 2
Sva finala igrana su u aprilu

```
pivotTeam = nba.select("season", "date_game").where((nba.is_playoffs == "Playoffs"))
pivotTeam.show()
```

```
from pyspark.sql.functions import max
```

```

pivotTeam = pivotTeam.groupBy("season").agg(max("date_game"))
pivotTeam.show()
```

Hipoteza nije tacna, vidi se da nisu sva finala odigrana u aprilu
